Notebook to create distance graph between frameworks (the graph was reconstructed in draw.io for the final thesis)

```python
"""
==============
Weighted Graph
==============

An example using Graph as a weighted network.
"""
import matplotlib.pyplot as plt
import networkx as nx

G = nx.Graph()

edges = [
('w', 'wc', 0.655),
('w', 'f1', 0.686),
('w', 'f3', 0.998),
('w', 'f4', 1.073),
('wc', 'f1', 0.262),
('wc', 'f3', 0.744),
('wc', 'f4', 0.708),
('f1', 'f3', 0.713),
('f1', 'f4', 0.693),
('f3', 'f4', 0.156),
]

for edge in edges:
    G.add_edge(edge[0], edge[1], weight=1.1-edge[2])





# G.add_edge("w", "f", weight=0.9)
# G.add_edge("a", "d", weight=0.3)

# elarge = [(u, v) for (u, v, d) in G.edges(data=True) if d["weight"] > 0.07]
# esmall = [(u, v) for (u, v, d) in G.edges(data=True) if d["weight"] <= 0.07]

pos = nx.spring_layout(G, seed=3)  # positions for all nodes - seed for reproducibility

# nodes
nx.draw_networkx_nodes(G, pos, node_size=700, node_shape ='o')

# edges
for edge in edges:
    nx.draw_networkx_edges(G, pos, edgelist=[(edge[0], edge[1])], width= (1.1 - edge[2]) * 2)

labels = {}
labels['w'] = r"$w_N$"
labels['wc'] = r"${wc}_{w_N}$"
labels['f1'] = r"$CN_{word1}$"
labels['f3'] = r"$CN_{word3}$"
labels['f4'] = r"$CN_{word4}$"

# labels
nx.draw_networkx_labels(G, pos, labels, font_size=20, font_family="sans-serif")

ax = plt.gca()
ax.margins(0.10)
plt.axis("off")
plt.tight_layout()
plt.show()
```
